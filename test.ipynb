```python
import requests

def get_location_by_ip():
    response = requests.get("https://ipinfo.io/json")
    data = response.json()
    lat, lon = data["loc"].split(",")
    return float(lat), float(lon)

lat, lon = get_location_by_ip()
print(f"Latitude: {lat}, Longitude: {lon}")
```
